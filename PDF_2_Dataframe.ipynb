```python
data=pd.DataFrame(columns=['date', "pdf_text",'Authors','Title'])
```

```python
#In case of multiple pdf files we can run the below loop and parse through them.
#Each of the pdf file will create a new row.

#for file in all_files:
#  raw = parser.from_file("/content/drive/My Drive/Toxicology_PDF/"+file)

pdf_text=raw['content']
date=raw["metadata"].get("date", "")
date=date[0:10]
authors = raw['metadata'].get('meta:author','')
Title = raw['metadata'].get('title','')
data=data.append({"date":date,"pdf_text":pdf_text,"Authors":authors,"Title":Title},ignore_index=True)
```

```python
data.head()
```

```python
data["pdf_text"].str.replace('[^A-Za-z\s]+','') #1st half
data["pdf_text"] = data["pdf_text"].str.replace('[^\w\s]','') #punctuations
data["pdf_text"] = data["pdf_text"].apply(lambda x: ' '.join([w for w in x.split() if len(w)>2]))
data.head()
```
